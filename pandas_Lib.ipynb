**SERIES**

```python
#create pandas series - Single Column
#lables - index by default start with 0
a=["Hi","I","am","Devaraj"]
b = pd.Series(a)
b
```

```python
#lables - custom indes
a=["Hi","I","am","Devaraj"]
b = pd.Series(a,index=['21','22','23','24'])
b
```

```python
#create pandas series by dictionary
data={"a":[1,2,3],"b":[4,5,6]}
b=pd.Series(data, index=['a','b'])
b
```

**DATAFRAME**

```python
#create pandas DataFrame by dictionary (default row index)
data={"a":[1,2,3],"b":[4,5,6]}
b=pd.DataFrame(data)
b
```

```python
#create pandas DataFrame by dictionary(custom row index)
data={"a":[1,2,3],"b":[4,5,6]}
b=pd.DataFrame(data, index=['a','b','c'])
b
```
